### KSD-Methode 


```python
def KSD(x,W,Sqx):
    x = x.reshape(-1,1)
    #h = np.sqrt(0.5 * ratio_median_heuristic(x,W,Sqx))
    #h = ratio_median_heuristic(x,W,Sqx)
    h = np.sqrt(0.5 * find_median_distance(x)) 
    print(h)
    Sqx = np.matrix(Sqx(x,W)).T  
    x = x.reshape(1,-1)
    x = np.matrix(x).T
    XY = x @ x.T
    
    x2 = np.multiply(x,x)
    sumx = np.multiply(x,Sqx)

    X2e = np.tile(x2, (1,np.shape(x)[0]))
    H = (X2e + X2e.T - 2*XY)
    Kxy = np.exp(-H/(2*h**2))
    
    sqxdy = -(Sqx @ x.T - np.tile(sumx,(1,np.shape(x)[0])))/h**2
    dxsqy = sqxdy.T
    dxdy = (-H/h**4 + 1/h**2)
    M = np.multiply((Sqx @ Sqx.T + sqxdy + dxsqy + dxdy) , Kxy)
    u = M - np.diag(np.diag(M))
    ksd = np.sum(u)/(np.shape(x)[0]*(np.shape(x)[0]-1))
    return u,ksd 

def boot(nboot,n,u,ksd):
    bootstrapSamples = np.zeros(nboot)
    for i in range(nboot):
        wtsboot = (stats.multinomial.rvs(n,size=1,p=1/n * np.ones(n))/n).T
        bootstrapSamples[i] = (wtsboot.T-1/n)@u@(wtsboot-1/n)
    p = np.mean(bootstrapSamples >= ksd)
    return p   
```

## Rejection Sampling

```python
def sample_x(num_samples,W_inv, xmin=-6, xmax=6, ymax=0.29, batch=1000):
    G = lambda x: np.exp(-2*np.log(np.cosh(np.pi/(2*np.sqrt(3)) * x )) - np.log(4))
    samples = []
    while len(samples) < num_samples:
        x = np.random.uniform(low=xmin, high=xmax, size=batch)
        y = np.random.uniform(low=0, high=ymax, size=batch)
        samples += x[y < G(x)].tolist()
        #print(np.shape(samples[:num_samples]))
    values = W_inv @ samples[:num_samples]
    return values.reshape(-1,1)
```

## Score Funktion
$\psi(x;W) =: s_q $

```python
s_q = lambda x,W: np.sum(W[:,i].reshape(-1,1) @ (-np.pi/3 * np.tanh(np.pi/(2*np.sqrt(3)) * (W[:,i] @x))) for i in range(np.shape(x)[0]))
```

```python
n = 1000
G = lambda x: np.exp(-2*np.log(np.cosh(np.pi/(2*np.sqrt(3)) * x )) - np.log(4))
x = np.linspace(-5,5,n)
W_k = np.eye(n)
W_inv_k =  np.linalg.inv(W_k)
W_f = np.eye(n) * 3
W_inv_f =  np.linalg.inv(W_f)
    

sample1 = sample_x(n,W_inv_k)
sample2 = sample_x(n,W_inv_f)
    
plt.hist(sample1, density=True, alpha = 0.5,bins = 50,label= '$W = I_n$')
plt.hist(sample2, density=True, alpha = 0.5,bins = 50, label='$W = 3* I_n$')
plt.plot(x, s_q(x.reshape(-1,1),W_k) , label = '$s_q(x,I_n)$')
plt.plot(x, s_q(x.reshape(-1,1),W_f), label = '$s_q(x,3 *I_n$)')
plt.plot(x, G(x), label = 'G')

plt.xlim(right=3)  # adjust the right leaving left unchanged
plt.xlim(left=-3)  # adjust the left leaving right unchanged
plt.legend()
#plt.savefig(fname = 'sample')
```

## Test

```python
def test_KSD(sample_size, s_q, sample_x,i,scale):
    print(i)
    error_rate = 0
    n = sample_size
    alpha = 0.05
    
    W = np.eye(n)
    W_inv =  np.linalg.inv(W)
    
    W_f = np.eye(n) * scale
    #W_inv_f =  np.linalg.inv(W_f)
    
    for i in range(1): 
        
        # 0 Richtige Verteilung; 1 Falsche Verteilung
        d = random.choice([0,1])
        if d == 0:
            x = sample_x(n,W_inv)
            u, ksd = KSD(x,W,s_q)

            p = boot(1000,sample_size,u,ksd)
            if p < alpha:
                error_rate = error_rate + 1
        else: 
            x =  sample_x(n,W_inv)
            u,ksd = KSD(x, W_f,s_q)
            p = boot(1000,sample_size,u,ksd)
            if p > alpha:
                error_rate = error_rate + 1
    return error_rate/1
```

```python
counter = 0
for scale in [0.1]:
    counter = counter +1
    sample_sizes = [20,50,80,200,500,1000]
    errors = []
    for i in range(len(sample_sizes)):
        error = test_KSD(sample_sizes[i],s_q, sample_x,i,scale)
        print(error)
        errors.append(error)

    plt.plot(sample_sizes,errors, 'b+-', label = scale)
    plt.legend()
    #plt.savefig(fname='test10 %s'%counter)
    plt.figure()
```
